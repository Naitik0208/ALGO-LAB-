##1.2 Arrays
Programs on 1-d arrays like - sum of elements of array, searching an element inarray, finding minimum and maximum element in array, count the number ofeven and odd numbers in array. For all such programs, also find the timeComplexity, compare if there are multiple methods

```python
def sum_arr(arr):
    total = 0
    for num in arr:
        total += num
    return total
# Time Complexity will be O(n)
arr_example = [1, 2, 3, 4, 5]
print("Sum of array:", sum_arr(arr_example))
# Output: 15
```

```python
#write a function named as max_min to find minimum (or maximum) element in an array
def max_min(arr, find="max"):
    if not arr:
        return None
    result = arr[0]
    for num in arr:
        if find == "max" and num > result:
            result = num
        elif find == "min" and num < result:
            result = num
    return result
# Time Complexity will be  O(n)
```

```python
# write a function named as Counting_EvenOdd to count number of even and odd elements in an array
def Counting_EvenOdd(arr):
    even_count = 0
    odd_count = 0

    for num in arr:
        if num % 2 == 0:
            even_count += 1
        else:
            odd_count += 1
    return even_count, odd_count
# Time Complexity will be O(n)
```

```python
# write a function named as linear_Search to perform a linear search in 1D-array
def linear_Search(arr,target):
    for index in range(len(arr)):
        if arr[index] == target:
            return index
    return -1
array = [10, 23, 45, 70, 11]
search_element = 45
result = linear_Search(array, search_element)
if result != -1:
    print(f"Element found at index {result}")
else:
    print("Element not found")
```
